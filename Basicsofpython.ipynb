Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
import math         #to use sqrt function

def is_prime(n):
  if n<=1:            #because it does not have 2 divisors
       return False
  if n==2:
     return True
  if n%2==0:
    return False
  if n>2:
    for i in range(3,int(math.sqrt(n))+1,2):   #to check divisibility up to sqrt(n) and int() to convert the float into int
      if n%i==0:
        return False
    return True

n=int(input("Enter any number:"))
if is_prime(n):
  print(n,"is prime")
else:
  print(n, "is not prime")
```

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random  #random is imported to generate random numbers
a=random.randint(1,100)
b=random.randint(1,100)
print(f"what is the product of {a},{b}?")   # prints the question in output screen
n=int(input("Enter the product of two number:")) # takes the product from user to check
if a*b==n:
  print("correct! Well Done")
else:
  print("Your answer is incorrect. The correct answer is:",a*b)  # Displays it is incorrect and provides answer
```

*	Had medicine charges of ₹2,400
