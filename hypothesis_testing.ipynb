```python
import numpy as np
import pandas as pd
from scipy.stats import t, norm

data=np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

df=pd.DataFrame(data)
alpha=0.01 # 99% confidence interval

df
```

```python
sample_mean=df.mean()   #sample mean
```

```python
sample_std=df.std() # sample standard deviation
```

```python
n=df.count()
degreeoffreedom=n-1  # degrees of freedom
```

```python
t_crit = t.ppf(1 - alpha/2,degreeoffreedom) # two-tailed t-test critical value because data is less than 30 values
se= sample_std / np.sqrt(n)    # standard error
CI_lower_t = sample_mean - t_crit * se #lower confidence interval
CI_upper_t = sample_mean + t_crit * se #upper confidence interval
```

```python
print(f"  99% Confidence Interval = ({CI_lower_t[0]}, {CI_upper_t[0]})")
```

two-tailed t-test because data is less than 30 values.two tailed because we have to find the lower and upper interval of confidence

```python
population_sigma = 0.2  # given population SD
z_crit = norm.ppf(1 - alpha/2)  # two-tailed z-test critical value
SE_z = population_sigma / np.sqrt(n)
```

```python
CI_lower_z = sample_mean - z_crit * SE_z   #lower confidence interval
CI_upper_z = sample_mean + z_crit * SE_z   #upper confidence interval
print(f"99% Confidence Interval = ({CI_lower_z[0]}, {CI_upper_z[0]})")
```

two-tailed z-test because the population standard deviation is given.two tailed because we have to find the lower and upper interval of confidence.
