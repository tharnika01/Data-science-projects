**Descriptive Analytics for Numerical Columns**

```python
import pandas as pd
```

```python
data=pd.read_csv('/content/sales_data_with_discounts.csv')
```

```python
data.head(3)
```

```python
data.info()
```

Volume, Avg Price, Total Sales Value, Discount Rate(%), Discount Amount, Net Sales Value are the numerical columns in the dataset

```python
data.describe()  #gives the mean, median, and standard deviation of columns.
```

```python
for col in data.columns:
  print(data[col].mode())   #gives mode of every column
```

* **volume**: mean is slightly greater
than median indicating a light skewness and max is much higher than 75% indicating only few large transactions.
* **Avg** **Price**:mean is greater than median indicating a positive skewness.
* **Total Sales Values**:mean is greater than median indicating a positive skewness.
* **Discount**: 25%, 50%, 75% have light difference, indicating discount range
* **Discount Amount**:mean is greater than median indicating a positive skewness.
* **Net Sales Value**: mean is greater than median indicating a positive skewness.
Avg Price,Total Sales Values,Discount Amount, Sales Value has outliers.
* **Mode**: Gives most repeated values. Some are Unimode,Bimodal and some are Multimodal

In conclusion, there are few large transactions and many normal transactions.

#catergorical data

```python
data['Date'].value_counts().plot.bar()
```

This bar plot shows that the sales occurred 30 times each day for the period from April 1st to April 15th, 2021. The height of each bar is consistent across all dates, indicating a constant daily frequency. There are no fluctuations in the sales.

```python
data['Day'].value_counts().plot.bar()
```

This bar plot shows that sales occurred each day of week. The height of each bar is consistent except Thursay, indicating a high frequency of sales compared to other days in week.

```python
data['SKU'].value_counts().plot.bar()
```

This bar plot shows all kind of the different SKU. The height of each bar is consistent across all unique codes of SKU(Stock Keeping Unit), indicating a constant frequency.


```python
data['City'].value_counts().plot.bar()
```

This bar plot shows that all the sales are done in City C.

```python
data['BU'].value_counts().plot.bar()
```

This bar plot shows the sales of Mobiles , FMCG and Lifestyle. The height of each bar is consistent, indicating a constant frequency. There are no fluctuations in the sales between them.

```python
data['Brand'].value_counts().plot.bar()
```

The bar plot is showing the frequency of sales of specific brand. 'Jeera' is the most frequent brand, followed by 'Viva' and 'Gear', while 'Orange' and 'Sumsang' are the least frequent.

```python
data['Model'].value_counts().plot.bar()
```

This bar plot shows that sales of each model. The height of each bar is consistent, indicating a high frequency of sales of each model.
