```python
from sklearn.preprocessing import OneHotEncoder
oe=OneHotEncoder()

gender=oe.fit_transform(df[['sex']]).toarray().astype('int')  # fit and transforms the data
encoded=pd.DataFrame(gender,columns=['Female','Male'])
encoded.head()
```

```python
final=df.join(encoded)
```

```python
final.drop('sex',axis=1,inplace=True)
```

```python
final.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in final.select_dtypes(include='object').columns:
    final[col] = le.fit_transform(final[col])

final
```

pros
* simple and fast
* no extra columns are created

cons
* not suitable for nominal catergory type of data

```python
def age_category(age):
    if age < 25:
        return 'Young'
    elif age < 45:
        return 'Middle-aged'
    elif age < 65:
        return 'Senior'
    else:
        return 'Elder'

df['age_group'] = df['age'].apply(age_category)
```

```python
def hours_category(hours):
    if hours <= 30:
        return 'Part-time'
    elif hours <= 40:
        return 'Full-time'
    elif hours <= 60:
        return 'Overtime'
    else:
        return 'Extreme'

df['hours_per_week_label'] = df['hours_per_week'].apply(hours_category)
```

```python
df
```

As income often varies by age group because the older people might earn more and hours_per_week_Label label the data and gives clear picture of the type of job the person is doing

```python
df[['capital_gain', 'capital_loss', 'hours_per_week', 'age','education_num']].skew()
```

Capital_gain and capital_loss are highly skewed and other columns does not No transformation needed

```python
df['capital_gain_log'] = np.log(df['capital_gain'] + 1)
df['capital_loss_log'] = np.log(df['capital_loss'] + 1)
df[['capital_gain_log', 'capital_loss_log']].skew()
```

still value is greater than +1.we should apply cube root

```python
df['cag_gain_cuberoot'] = np.sqrt(df['capital_gain_log'])
df['cag_loss_cuberoot'] = np.sqrt(df['capital_loss_log'])
df[['cag_gain_cuberoot', 'cag_loss_cuberoot']].skew()
```

even after this transformation the skewness has not decreased.so, either have to drop the columns as it contains many zeroes in the column
